```python
import requests
from bs4 import BeautifulSoup
```

```python
res   = requests.get('https://en.wikipedia.org/wiki/Mahatma_Gandhi')
soup  = BeautifulSoup(res.text,'html.parser')
```

```python
soup.text
```

```python
heading = soup.find('h1').text
```

```python
heading
```

```python
corpus = ''
for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'

corpus = corpus.strip()
```

```python
print(corpus)
```

```python
fd = open(heading + '.txt', 'w', encoding='utf-8')
fd.write(corpus)
fd.close()
```
