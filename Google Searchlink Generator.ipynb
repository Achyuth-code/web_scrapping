```python
res  = requests.get('https://en.wikipedia.org/wiki/Mahatma_Gandhi')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup
```

```python
heading = soup.find('h1').text
```

```python
heading
```

```python
corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'

corpus = corpus.strip()
for i in range(500):
    corpus = corpus.replace('['+str(i)+']','')
```
